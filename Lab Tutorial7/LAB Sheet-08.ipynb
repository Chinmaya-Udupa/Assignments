## 1. Training a Decision Tree
We'll use the pre-loaded [Iris](https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html) data set and build a [Decision Tree Classifier](https://scikit-learn.org/stable/modules/tree.html).

Start by loading the dataset and building a decision tree classifier.

(hint: Use $\texttt{DecisionTreeClassifier}$ model from $\texttt{sklearn.tree}$)

```python
from sklearn import tree
from sklearn import datasets
```

```python
# Your code solution here
iris = datasets.load_iris()
X = iris.data
y = iris.target
clf = tree.DecisionTreeClassifier(max_depth=2)
```
